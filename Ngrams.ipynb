```python
def bigram(corpus):
    """
    This takes a corpus, cleans it of any punctiation, using the string punctiation package and splits it into a oekenzed form
    it then makes a tuple dict out of the rsult which can be used for analysis
    """
    cleaned_corpus = ''.join(word for word in corpus if word not in punctuation).split(" ") # Corpus Cleaner
    bigram_dict = dict() # Dict to be storing the bigram
    
    for i in range(1,len(cleaned_corpus)): # loop over all of the words starting at position1 ( to look back)
        if (cleaned_corpus[i-1],cleaned_corpus[i]) not in bigram_dict: # if the  work sint in it, then isr it on the enxt line
            bigram_dict[(cleaned_corpus[i-1],cleaned_corpus[i])] = 0
        bigram_dict[(cleaned_corpus[i-1],cleaned_corpus[i])] += 1 # increment the count
    print(bigram_dict)
    
    
# Awesome works as expected
bigram(corpus)
```
